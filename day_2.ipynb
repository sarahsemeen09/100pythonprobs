Write a program which accepts a sequence of comma-separated numbers from console and generate a list and a tuple which contains every number.

Suppose the following input is supplied to the program: 34,67,55,33,12,98

Then, the output should be:
['34', '67', '55', '33', '12', '98']

```
def get_list(n_list):
  result = []
  for each in n_list:
    result.append(each)
  print(result)
```

```
if __name__ == "__main__":
  given = input("Please enter a list of numbers, comma seperated").split(",")
  get_list(given)
```

```
#or simply 
given = input("Please enter a list of numbers, comma seperated").split(",")
print(given)
```

Write a program that calculates and prints the value according to the given formula:

Q = Square root of [(2 _ C _ D)/H]

Following are the fixed values of C and H:

C is 50. H is 30.

D is the variable whose values should be input to your program in a comma-separated sequence.

For example Let us assume the following comma separated input sequence is given to the program:
100,150,180

The output of the program should be:
18,22,24

```
import math

def sqrt_root(n_list):
  c=50
  h=30
  result = []
  for each in n_list:
    answer=round(math.sqrt((2*c*int(each))/h))
    result.append(answer)
  print(result)
```

```
if __name__ == "__main__":
  given = input("Please enter a list of numbers, comma seperated").split(",")
  sqrt_root(given)
```

_Write a program which takes 2 digits, X,Y as input and generates a 2-dimensional array. The element value in the i-th row and j-th column of the array should be i * j.

Note: i=0,1.., X-1; j=0,1,¡­Y-1. 

Suppose the following inputs are given to the program: 3,5

Then, the output of the program should be:
[[0, 0, 0, 0, 0], [0, 1, 2, 3, 4], [0, 2, 4, 6, 8]]

```
given = input("Give the dimensions").split(",")
n1 = int(given[0])
n2 = int(given[1])

result = []
for i in range(0,n1):
  inside = []
  for j in range(0,n2):
    inside.append(i*j)
  result.append(inside)
print(result)
```

Write a program that accepts sequence of lines as input and prints the lines after making all characters in the sentence capitalized.

Suppose the following input is supplied to the program:

Hello world\
Practice makes perfect

Then, the output should be:

HELLO WORLD\
PRACTICE MAKES PERFECT

```
given = input("Say something").split(" ")
result = []
for each in given:
  result.append(each.upper())
print(*result)
```
